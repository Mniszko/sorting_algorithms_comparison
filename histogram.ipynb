```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dane = pd.read_csv('./can.txt',delimiter='\t',decimal=',')

def min(arr,T):
    min=1e20
    for i in arr:
        if i<min and i!=0:
            min=i
        if T and i<min:
            min=i
    return min

def max(arr):
    max=0
    for i in arr:
        if i>max:
            max=i
    return max
#"selection","insertion","bubble","quick","merge","heap"
```

```python
n=1
no1 = dane.iloc[n-1:n,].values
no2 = dane.iloc[n:n+1,].values
arr=[]
fig,ax = plt.subplots(1,2)
fig.set_size_inches(18, 5, forward=True)

for i in no1:
    arr.append(i)
arr2=np.array(arr)[0]

ax[0].hist(arr2,bins=15,range=[min(arr2,False),max(arr2)],rwidth=0.9)
ax[0].title.set_text('selection (ingerencje w listę)')

arr=[]
for i in no2:
    arr.append(i)
arr2=np.array(arr)[0]

ax[1].hist(arr2,bins=15,range=[min(arr2,False),max(arr2)],rwidth=0.9)
ax[1].title.set_text('selection (porównania)')
plt.savefig('num'+str(n)+'.png',bbox_inches="tight")
```
